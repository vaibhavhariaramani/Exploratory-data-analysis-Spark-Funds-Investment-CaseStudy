# Checkpoint 1

### Data Cleaning

```python
import pandas as pd

#Reading a data frame from a text file
companies = df=pd.read_table('companies.txt',header=0, encoding="ISO-8859-1")

#Converting all the permalink strings to lower case  --> Needed for merging
companies.permalink = companies.permalink.str.lower()
companies
```

```python
#Reading the data frame present in rounds2.csv file
rounds2 = pd.read_csv("rounds2.csv", encoding="ISO-8859-1")

#Converting all the company_permalink strings to lower case  --> Needed for merging
rounds2.company_permalink = rounds2.company_permalink.str.lower()
rounds2
```

```python
#Reading mapping.csv into a data frame mapping
mapping = pd.read_csv('mapping.csv')
mapping.head()
```

```python
#Dropping the 0th index row which has just NaN value
mapping = mapping.drop(0,axis = 0)

#Dropping the Blanks column
mapping = mapping.drop(['Blanks'],axis = 1)
mapping.head()
```

```python
#Using pd.melt() with id variable category_list will map category_list element with all other columns in separate rows
mapping = pd.melt(mapping,id_vars=['category_list'],var_name = 'main_sector')

#Now we only need the keep the rows which have value 1 associated with them
mapping = mapping[mapping.values == 1]

mapping.head()
```

```python
#Renaming column names from category_list -> primary_sector and   variable -> main_sector
mapping = mapping.rename({'category_list':'primary_sector'})

#Dropping the value column
mapping = mapping.drop('value',axis = 1)

#Resetting indexes from 0 to length of mapping
mapping.reset_index(inplace = True)

#Dropping the index
mapping = mapping.drop('index',axis = 1)
mapping.head()
```

```python
#Renaming company_permalink column to permalink.
#This is required for merging as column name in companies data frame is permalink
rounds2.rename(columns = {'company_permalink':'permalink'},inplace = True)
rounds2
```

```python
#Merging the two data frames  together and storing in a new data frame master_df
#We are only taking the inner join of the required columns that are needed for further analysis
master_df = pd.merge(companies[['permalink','name','category_list','status','country_code']],rounds2[['permalink','funding_round_type','raised_amount_usd']],how='inner',on='permalink')
master_df
```

```python
#Checking the number of missing elements in each column
master_df.isnull().sum()
```

```python
#Checking the whether a column has missing elements or not
master_df.isnull().any()
```

```python
#No of elements missing in each row
master_df.isnull().sum(axis=1)
```

```python
#Removing all rows which have NaN values in raised_amount_usd
master_df = master_df[master_df.raised_amount_usd.isnull() == False]
master_df
```

```python
#Unique Companies in rounds2
print('Unique companies in rounds2 are',len(rounds2['permalink'].unique()))

#unique companies in companies
print('Unique companies in companies are',len(companies['permalink'].unique()))
```

# Checkpoint 2

### Funding type Analysis

```python
#Grouping all the values in master_df acc to the funding round type
round_df = master_df[master_df['funding_round_type'].isin(['seed','venture','angel','private_equity'])]
round_type_df = round_df.groupby('funding_round_type')
```

```python
#Finding the sum of raised_amount which is grouped by round type
#Then converting it to a data frame and sorting acc to the raised amount -> for improving readability
pd.DataFrame(round_type_df['raised_amount_usd'].sum()).sort_values(by='raised_amount_usd')
```

```python
#Now for finding the investment per round of a round type, we will simply take average/mean
#Then convert it to a data frame and sort by raised amount
per_round_amount = pd.DataFrame(round_type_df['raised_amount_usd'].mean()).sort_values(by='raised_amount_usd')
per_round_amount
```

#### 2.1 Average funding amount of venture type is 1.174895e+07 ie 11.74 million USD

#### 2.2 Average funding amount of angel type is 9.586945e+05 ie 0.95 million USD

#### 2.3 Average funding amount of seed type is 7.197796e+05 ie 0.71 million USD

#### 2.4 Average funding amount of private equity type is 7.330859e+07 ie 73.30 million USD

#### 2.5 venture --> Since only the per round average of venture lies between 5 and 15 million USD, it is the chosen investment type

# Checkpoint 4

### Sector Analysis 1

```python
#Removing all the rows which have NaN values in category_list
master_df = master_df[master_df.category_list.isnull() == False]

master_df.head()
```

```python
#Function to retrive the 1st category from the category list
def retrive(value):
    values=value.split('|')
    if(len(values)>0):
        return values[0]
    else:
        return value
```

```python
#Applying the above function to the category lists to obtain the primary sector
master_df['primary_sector'] = master_df['category_list'].apply(retrive)

master_df.head()
```

```python
master_df = pd.merge(master_df,mapping,how = 'inner',on = 'category_list')

master_df.head()
```

### A plot showing the fraction of total investments (globally) in venture, seed, and private equity, and the average amount of investment in each funding type

```python
graph1 = sns.barplot(x = 'funding_round_type', y = 'raised_amount_usd', data = round_df)
graph1.set_ylim(0, 25000000)
graph1.set(xlabel = 'Funding Round Type', ylabel = 'Raised Amount')
plt.show()
```

#### From this graph we can see that the value of venture lies between 1.0 and 1.5 x 10^7 ie 10-15 million USD. Venture is the only funding round type that exists in between 5 and 15 million USD. Hence it is the preffered Funding Type

### A plot showing the top 9 countries against the total amount of investments of funding type FT.
